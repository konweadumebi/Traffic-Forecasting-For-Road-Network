Plotting the difference between the routing system and without the routing system

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
routes = ['N1-N2', 'N2-N1', 'N2-N3', 'N3-N2', 'N3-N4', 'N4-N3', 'N4-N5', 'N5-N4', 'N5-N6', 'N6-N5']
total_time_with_system = [20, 18, 45, 42, 25, 23, 15, 17, 20, 18]
total_time_without_system = [25, 27, 58, 60, 40, 38, 30, 38, 30, 35]

# Create bar width and position
bar_width = 0.35
index = np.arange(len(routes))

# Plotting the bars
plt.bar(index, total_time_without_system, bar_width, label='without routing system', color='skyblue')
plt.bar(index + bar_width, total_time_with_system, bar_width, label='with routing system', color='gold')

# Adding labels and title
plt.xlabel('Routes')
plt.ylabel('Total Time')
plt.title('Comparison of Forecasting System on Total Time')
plt.xticks(index + bar_width / 2, range(1, 11))  # X-axis label positions
plt.ylim(5, 65)  # Adjusting the Y-axis limits to match the graph in the image
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
routes = ['N1-N2', 'N2-N1', 'N2-N3', 'N3-N2', 'N3-N4', 'N4-N3', 'N4-N5', 'N5-N4', 'N5-N6', 'N6-N5']
total_time_with_system = [20, 18, 45, 42, 25, 23, 15, 17, 20, 18]
total_time_without_system = [25, 27, 58, 60, 40, 38, 30, 38, 30, 35]

# Calculate the time gained (difference)
time_gained = [without - with_ for without, with_ in zip(total_time_without_system, total_time_with_system)]

# Create bar width and position
bar_width = 0.35
index = np.arange(len(routes))

# Plotting the bars
plt.bar(index, time_gained, bar_width, color='lightgreen')

# Adding labels and title
plt.xlabel('Routes')
plt.ylabel('Time Gained (without - with)')
plt.title('Time Gained by Using Routing System')
plt.xticks(index, range(1, 11))  # X-axis label positions
plt.ylim(0, 30)  # Adjusting the Y-axis limits to show positive differences clearly

# Show the plot
plt.show()
```
